```python
#add in cost of goods
cog = [0.00, 22.96, 130.11, 147.03, 596.66, 629.24, 827.98, 518.91, 566.59, 4.30, 4.30]
df_totals['Cost of Goods'] = cog
df_totals
```

```python
df_totals.info()
```

```python
# create data
x = df_totals['Pay Date'] 
y1 = df_totals['Payment'] 
y2 = df_totals['Cost of Goods'] 
 
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.title('2022 Monthly Payments and Expenses')
# show legend
plt.legend(handles = [y1, y2], 
           labels  = ['Payments', 'Goods'])
plt.show()
```

```python
#add in cost of supplies and a calculated column
supp = [37.32, 185.17, 186.38, 286.48, 151.91, 0, 116.31, 100.02, 81.53, 94.04, 9.99]
df_totals['Supplies'] = supp
df_totals['Total Expenses'] = df_totals.loc[:,['Cost of Goods','Supplies']].sum(axis=1)
df_totals
```

```python
# Data

x = df_totals['Pay Date'] 
y1 = df_totals['Payment'] 
y2 = df_totals['Total Expenses'] 

X_axis = np.arange(len(x))
payment_bar=plt.bar(X_axis - 0.2,y1,0.4)
expenses_bar=plt.bar(X_axis + 0.2,y2,0.4)
plt.title = "2022 Reseller Payments and Expenses"
plt.xticks(X_axis, x)
plt.legend(handles = [payment_bar, expenses_bar], 
           labels  = ['Payment', 'Total Expenses'])
plt.show()
```
